```python
import pandas as pd #importing pandas library
```

```python
ger = pd.read_csv('players_20.csv')  #importing our dataset and storing it in a variable ger
ger
```

```python
ger1 = ger[(ger.nationality == "Germany")] #selecting only german players from the dataset
ger1
```

```python
ger2 = ger1[(ger1.team_position == "GK")]  #displaying german goalkeepers and storing it in variable ger2
ger2
```

```python
ger3 = ger2.sort_values(['gk_diving', 'gk_handling' ,'gk_reflexes'], ascending=[False,False,False]) #finding best german goalkeeper on the basis of attributes such as ball handling,reflexes and diving
ger3.head(5) #storing data in variable ger3, arranging data in descending order 
```
